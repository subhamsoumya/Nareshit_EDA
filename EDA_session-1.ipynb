# Method-1:

**Create dataframe using list**

```python
Names=['Ramesh','Suresh','Sathish']
Age=[20,25,30]
pd.DataFrame(zip(Names,Age))
```

**Update the column names**

```python
Names=['Ramesh','Suresh','Sathish']
Age=[20,25,30]
pd.DataFrame(zip(Names,Age),columns=['Name','Age'])
```

**Update index/column names**

```python
Names=['Ramesh','Suresh','Sathish']
Age=[20,25,30]
df1=pd.DataFrame(zip(Names,Age),
             columns=['Name','Age'],
             index=['A','B','C'])
df1
```

**Change entire column data**

```python
df1['Name']=['Sathish','Suresh','Ramesh']
df1
```

**Create a new column**

```python
df1['City']=['Hyderabad','Bangalore','Chennai']
df1
```

**Changing entire row data**

```python
df1.loc['C']=['Soumya',28,'Odisha']
df1
```

```python
# CHange the index A B C to 0 1 2
Names=['Ramesh','Suresh','Sathish']
Age=[20,25,30]
df1=pd.DataFrame(zip(Names,Age),
             columns=['Name','Age'],
             index=[0,1,2])
df1
```

```python
len(df1.index)
```

**Add new rows**

```python
df1.loc[3]=['Alex',35]
df1
```

```python
df1.loc['A']=['Bunny',40]
df1
```

**Drop the column**

- Before  drop the column

- I will create a new column

- I have only two columns are there Name and Age

- I will create a city. Then i will drop it

```python
Names=['Ramesh','Suresh','Sathish']
Age=[20,25,30]
df1=pd.DataFrame(zip(Names,Age),columns=['Name','Age'])
df1['City']=["Hyd",'Blr','Chennai']
df1
```

```python
# Axis=1  It reference Column
# Axis=0  It reference Rows
# The result you want to overwrite or save in another variable
# If you want to overwrite in same dataframe , you need to use inplace=True

df1.drop('City',axis=1,inplace=True)
```

```python
df1
```

**Drop the row**

```python
df1.drop(2,inplace=True)
df1
```

**saving output in another folder**

```python


df1.to_csv("output.csv",index=False)
```
