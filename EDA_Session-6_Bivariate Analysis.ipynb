```python
# I want to know out of 16861 members how many got the visa
# we are analysing continent and case status
```

```python
con1=visa_df['continent']=='Asia'
con2=visa_df['case_status']=='Certified'
df1=visa_df[con1&con2]
df1
```

```python
con1=visa_df['continent']=='Asia'
con2=visa_df['case_status']=='Denied'
df1=visa_df[con1&con2]
df1
```

```python
con1=visa_df['continent']=='Asia'
con2=visa_df['case_status']=='Certified'
con3=visa_df['case_status']=='Denied'
cert_con=con1&con2
denied_con=con1&con3
len(visa_df[cert_con]),len(visa_df[denied_con])
print(f"{len(visa_df[cert_con])}  got the Visa from Asia")
print(f"{len(visa_df[denied_con])} got rejected the Visa from Asia")
```

```python
labels=visa_df['continent'].value_counts().keys()
for i in labels:
    con1=visa_df['continent']==i
    con2=visa_df['case_status']=='Certified'
    con3=visa_df['case_status']=='Denied'
    cert_con=con1&con2
    denied_con=con1&con3
    len(visa_df[cert_con]),len(visa_df[denied_con])
    print(f"{len(visa_df[cert_con])}  got the Visa from {i}")
    print(f"{len(visa_df[denied_con])} got rejected the Visa from {i}")
```

```python
labels=visa_df['continent'].value_counts().keys()
certified=[]
denied=[]
for i in labels:
    con1=visa_df['continent']==i
    con2=visa_df['case_status']=='Certified'
    con3=visa_df['case_status']=='Denied'
    cert_con=con1&con2
    denied_con=con1&con3
    certified.append(len(visa_df[cert_con]))
    denied.append(len(visa_df[denied_con]))

certified,denied
```

```python
pd.DataFrame(zip(labels,certified,denied),columns=['Continent','Certified','Denied'])
```

```python
d1=pd.DataFrame(zip(labels,certified,denied),columns=['Continent','Certified','Denied'])
d1.set_index('Continent')
```

**Numerical vs Numerical**

**Scatter-plots:**
    
   - plt.scatter
    
   - will take two arguments x-axis and y-axis
    
   - Both variables should be numerical
    
   - It provides relationship between two variables
    
        - Positively related
        
        - Negatively related
        
        - No relation

```python
x=[1,2,3,4,5]
y=[11,12,13,14,15]
plt.scatter(x,y)
plt.plot(x,y)
```

```python
x=[i for i in range (-4,5)]
y=[i*i for i in x]
plt.scatter(x,y)
plt.plot(x,y)
```

```python
col1=visa_df['prevailing_wage']
col2=visa_df['no_of_employees']
plt.scatter(col1,col2)
#plt.plot(col1,col2)
```

**Corelation-Coefficient**

   - Denoted with r
    
   - Pearson corelation coefficient
        
   - r varies from -1 to 1
    
   - r= -1 to 0 : Negatively corelated
            
   - r= 0 to 1 : Positively corelated
        
   - r=0 : No relation

```python
visa_df.corr(numeric_only=True)

# Imagine you have 10 Column 10*10=100
```

```python
x1=visa_df['prevailing_wage']
y1=visa_df['yr_of_estab']
plt.scatter(x1,y1)
```
