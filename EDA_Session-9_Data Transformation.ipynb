```python
import numpy as np
import matplotlib.pyplot as plt
```

**Exponential-data**

```python
exp_data=np.random.exponential(size=1000)
exp_data[:10]

# We are considering exponential data set with random 1k observations
# using numpy package
```

```python
plt.hist(exp_data,bins=30,label='Exponential')
plt.legend()
plt.show()
```

**Norm-data**

```python
norm_data=np.random.normal(size=1000)
plt.hist(norm_data)
```

**Log Transformation**

- np.log is used for log transformation

- Generally log transformation will not convert data into normal

- It avoids skewness

- np.log means natural logarithm , base=e

```python
log_data=np.log(exp_data)
log_data[:10]
```

```python
exp_data[:10]
```

```python
plt.hist(log_data,bins=40,label='Log data')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,
                        bins=40,
                        label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(log_data,
                        bins=40,
                        label='Log Transformation')
plt.legend()
plt.show()
```

**Reciprocal transformation**

- Reciprocal transformation fails when data has zero value

```python
rec_data=np.reciprocal(exp_data)
plt.hist(rec_data,bins=40,label='Reciprocal data')
plt.legend()
plt.show()
```

**Square root transformation**

```python
print(25**2) # square
print(25**(1/2)) # square root
```

```python
print(np.sqrt(25))
```

```python
sqrt_data=np.sqrt(exp_data)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,
                        bins=40,
                        label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(sqrt_data,
                        bins=40,
                        label='Square root Transformation')
plt.legend()
plt.show()
```

```python
import seaborn as sns
sns.displot(sqrt_data)
```

**Power transformation**

- It is related to sklearn package

- Package name: sklearn.preprocessing

- Method name: Power Transformer

- Inside Box-Cox , yeo-jhonson

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer()
power_data=pt.fit([exp_data]).transform([exp_data])
power_data
```

```python
plt.hist(power_data[0])
plt.show()
```
