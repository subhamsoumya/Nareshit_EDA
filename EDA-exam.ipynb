# 4.

```python
dict1=tele_df['Gender'].value_counts()
dtype=dict(dict1)
dtype
```

```python
dict1=dict(tele_df['Gender'].value_counts())
keys=list(dict1.keys())
values=list(dict1.values())
dict2=pd.DataFrame(zip(keys,values),columns=['Gender','Count'])
dict2.to_csv('Gender_df.csv',index=False)
```

```python
pd.read_csv('Gender_df.csv')
```

```python
plt.figure(figsize=(8,6))
plt.bar('Gender','Count',data=dict2)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Bar chart")
plt.savefig("Gender_bar_chart.jpg")
plt.show()  
```
